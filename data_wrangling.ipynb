```python
# remove # Samples	columns from carlino_supp
carlino_supp_updated = carlino_supp.drop('# Samples', axis=1)

# remove rows with dataset_name that are in missing_refs
carlino_supp_updated = carlino_supp_updated[~carlino_supp_updated['Dataset'].isin(missing_refs['dataset_name'])]

# Adjust column names in missing_refs to match carlino_supp
missing_refs = missing_refs.rename(columns={'dataset_name': 'Dataset', 'project_accession': 'Accession number'})


# concatenate carlino_supp and missing_refs
carlino_supp_updated = pd.concat([carlino_supp_updated, missing_refs], ignore_index=True)

# remove rows missing references
carlino_supp_updated = carlino_supp_updated.dropna(subset=['Reference'])

# check if the number of rows is the same as the sum of the two dataframes minus the duplicates
print(f"Number of rows in the concatenated dataframe: {carlino_supp_updated.shape[0]}")
print(f"Sum of the two dataframes: {carlino_supp.shape[0] + missing_refs.shape[0] - 4}")
```

```python
# Split rows with multiple bioproject accession numbers in carlino_supp
carlino_supp_updated['Accession number'] = carlino_supp_updated['Accession number'].str.split(',')
carlino_supp_updated = carlino_supp_updated.explode('Accession number')
carlino_supp_updated['Accession number'] = carlino_supp_updated['Accession number'].str.strip()

# Create a mapping dictionary using both Dataset and Accession number as keys
reference_dict = {}
for _, row in carlino_supp_updated.iterrows():
    # Map by dataset name
    reference_dict[('dataset', row['Dataset'])] = row['Reference']
    # Map by accession if available
    if pd.notna(row['Accession number']):
        reference_dict[('accession', row['Accession number'])] = row['Reference']

# Add Reference column to carlino_metadata using both dataset_name and project_accession
def get_reference(row):
    # Try to find reference by dataset name first
    ref = reference_dict.get(('dataset', row['dataset_name']))
    if pd.isna(ref) and pd.notna(row['project_accession']):
        # If not found, try to find by project accession
        ref = reference_dict.get(('accession', row['project_accession']))
    return ref

carlino_metadata['Reference'] = carlino_metadata.apply(get_reference, axis=1)

# Verify the merge by checking if References were added
print("Number of rows with References:", carlino_metadata['Reference'].notna().sum())
print("Total rows in carlino_metadata:", len(carlino_metadata))
```

Since Carlino is by far the largest dataset we have, I will adjust the other smaller sets to match it as much as possible. <br>

```python
import glob

# Get all files matching the pattern
files = glob.glob('data/Saak2023_CheeseMAGs_bin*')

# Function to merge two rows into one
def merge_rows(row1, row2):
    merged = {}
    for col in row1.index:
        # If values are the same, keep one
        if row1[col] == row2[col]:
            merged[col] = row1[col]
        # If values are different and not NaN, combine them
        elif pd.notna(row1[col]) and pd.notna(row2[col]):
            merged[col] = f"{row1[col]}; {row2[col]}"
        # If one value is NaN, keep the non-NaN value
        else:
            merged[col] = row1[col] if pd.notna(row1[col]) else row2[col]
    return pd.Series(merged)

# List to store all combined dataframes
all_dfs = []

# Process each file
for file in files:
    # Read the file
    df = pd.read_csv(file)
    
    # If there's only one row, add it directly
    if len(df) == 1:
        all_dfs.append(df)
    # If there are multiple rows, merge them
    else:
        merged_row = df.iloc[0]
        for i in range(1, len(df)):
            merged_row = merge_rows(merged_row, df.iloc[i])
        all_dfs.append(pd.DataFrame([merged_row]))

# Combine all dataframes
Saak2023_metadata = pd.concat(all_dfs, ignore_index=True)

# Display result
display(Saak2023_metadata)
```

```python
# loading the Rappaport isolate metadata
def parse_sample_data(file_path):
    # Initialize variables
    samples = []
    current_sample = {}
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
        
        for line in lines:
            line = line.strip()
            
            # Skip empty lines
            if not line:
                # If we have a sample, add it to our list and start a new one
                if current_sample:
                    samples.append(current_sample)
                    current_sample = {}
                continue
            
            # Check if this is a new sample header
            if line[0].isdigit() and '_' in line:
                if current_sample:  # Save previous sample if exists
                    samples.append(current_sample)
                current_sample = {'Sample_ID': line.strip()}
                continue
            
            # Parse key-value pairs
            if ':' in line:
                key, value = [x.strip() for x in line.split(':', 1)]
                current_sample[key] = value
    
    # Add the last sample if it exists
    if current_sample:
        samples.append(current_sample)
    
    # Convert to DataFrame
    df = pd.DataFrame(samples)
    
    return df

# Parse the file
rappaport_metadata = parse_sample_data('data/Rappaport2024_sourdough_biosample_result.txt')
display(rappaport_metadata)
```

```python
# striping the [1-9]: from the Sample_ID column
rappaport_metadata['Sample_ID'] = rappaport_metadata['Sample_ID'].str.replace(r'^\d+:\s+', '', regex=True)
display(rappaport_metadata)
```
