```python
carlino_metadata = pd.read_csv('data/cFMD_metadata.tsv', sep='\t')
carlino_supp = pd.read_excel('data/Carlino_supp.xlsx') #requires openpyxl to be installed
missing_refs = pd.read_csv('data/missing_references.tsv', sep='\t')

display(carlino_metadata.head())
display(carlino_supp.head())    
display(missing_refs.head())
```

```python
# checking for cross matches bewteen the missing references and the carlino_supp
display(missing_refs['dataset_name'].apply(lambda x: carlino_supp['Dataset'].str.contains(x).any()).sum() )

# display the rows that have a match in both datasets
display(missing_refs[missing_refs['dataset_name'].apply(lambda x: carlino_supp['Dataset'].str.contains(x).any())])
display(carlino_supp[carlino_supp['Dataset'].apply(lambda x: missing_refs['dataset_name'].str.contains(x).any())])

```

The Accession number in carlino_spp does not match the correct column in carlino_metadata. <br>
To ensure proper mapping, I will remove the rows from carlino_supp before concatenating. 

```python
# remove # Samples	columns from carlino_supp
carlino_supp_updated = carlino_supp.drop('# Samples', axis=1)

# remove rows with dataset_name that are in missing_refs
carlino_supp_updated = carlino_supp_updated[~carlino_supp_updated['Dataset'].isin(missing_refs['dataset_name'])]

# Adjust column names in missing_refs to match carlino_supp
missing_refs = missing_refs.rename(columns={'dataset_name': 'Dataset', 'project_accession': 'Accession number'})


# concatenate carlino_supp and missing_refs
carlino_supp_updated = pd.concat([carlino_supp_updated, missing_refs], ignore_index=True)

# remove rows missing references
carlino_supp_updated = carlino_supp_updated.dropna(subset=['Reference'])

# check if the number of rows is the same as the sum of the two dataframes minus the duplicates
print(f"Number of rows in the concatenated dataframe: {carlino_supp_updated.shape[0]}")
print(f"Sum of the two dataframes: {carlino_supp.shape[0] + missing_refs.shape[0] - 4}")
```

```python
# Split rows with multiple bioproject accession numbers in carlino_supp
carlino_supp_updated['Accession number'] = carlino_supp_updated['Accession number'].str.split(',')
carlino_supp_updated = carlino_supp_updated.explode('Accession number')
carlino_supp_updated['Accession number'] = carlino_supp_updated['Accession number'].str.strip()

# Create a mapping dictionary using both Dataset and Accession number as keys
reference_dict = {}
for _, row in carlino_supp_updated.iterrows():
    # Map by dataset name
    reference_dict[('dataset', row['Dataset'])] = row['Reference']
    # Map by accession if available
    if pd.notna(row['Accession number']):
        reference_dict[('accession', row['Accession number'])] = row['Reference']

# Add Reference column to carlino_metadata using both dataset_name and project_accession
def get_reference(row):
    # Try to find reference by dataset name first
    ref = reference_dict.get(('dataset', row['dataset_name']))
    if pd.isna(ref) and pd.notna(row['project_accession']):
        # If not found, try to find by project accession
        ref = reference_dict.get(('accession', row['project_accession']))
    return ref

carlino_metadata['Reference'] = carlino_metadata.apply(get_reference, axis=1)

# Verify the merge by checking if References were added
print("Number of rows with References:", carlino_metadata['Reference'].notna().sum())
print("Total rows in carlino_metadata:", len(carlino_metadata))
```

```python
# checking the missing references again
missing_refs = carlino_metadata[
    (carlino_metadata.Reference.isna()) & 
    (carlino_metadata['fermented/non-fermented'] == 'F')
][['dataset_name', 'project_accession','Reference']].drop_duplicates()

# check if the dataset_name is in the Fermentos metadata
missing_refs = missing_refs[missing_refs['dataset_name'].apply(lambda x: dataset['mag_id'].str.contains(x).any())]
display(missing_refs)
```
