# Data wrangling and curation of Fermentos dataset

```python
import pandas as pd # type: ignore
```

```python
# loading Fermentos metadata file
dataset = pd.read_csv('data/2025-03-24-all-ff-mag-metadata-cleaned-curated.tsv', sep='\t')
display(dataset.head())
```

```python
# loading the Carlino metadata. 
carlino_metadata = pd.read_csv('data/cFMD_metadata.tsv', sep='\t')
carlino_supp = pd.read_excel('data/Carlino_supp.xlsx') #requires openpyxl to be installed

display(carlino_metadata.head())
display(carlino_supp.head())
```

```python
# organising carlino_supp
    # replacing * with the Carlino paper
    # spliting rows with multiple bioproject accession numbers

# Replacing '*' in the Reference column with the full citation of the Carlino paper
ref = 'Carlino, N., Blanco-Míguez, A., Punčochář, M., Mengoni, C., Pinto, F., Tatti, A., Manghi, P., Armanini, F., Avagliano, M., Barcenilla, C., Breselge, S., Cabrera-Rubio, R., Calvete-Torre, I., Coakley, M., Cobo-Díaz, J. F., De Filippis, F., Dey, H., Leech, J., Klaassens, E. S., Knobloch, S., O’Neil, D., Quijada, N. M., Sabater, C., Skírnisdóttir, S., Valentino, V., Walsh, L., MASTER EU Consortium, Asnicar, F., Fackelmann, G., Heidrich, V., Rota Stabelli, O., Pasolli, E., Arendt, E., Rubino, F., Cotter, P. D., Ruiz, L., Ercolini, D., Morrison, S., Marchart, J., Sessitsch, A., and Zannini, E. (2024). A pan-European study of the human gut microbiome reveals links with diet, health, and lifestyle. Cell, 187(13), 5795–5816.e15. https://doi.org/10.1016/j.cell.2024.05.028'
carlino_supp.loc[carlino_supp['Reference'] == '*', 'Reference'] = ref

# Splitting rows with multiple bioproject accession numbers
# check how many rows have multiple Bioproject accession numbers
# Find rows with commas in 'Accession number' and count values after splitting
rows_with_commas = carlino_supp[carlino_supp['Accession number'].str.contains(',', na=False)]
total_values = sum(len(accession.split(',')) for accession in rows_with_commas['Accession number'])
print(f"Total values after splitting: {total_values}")

print("\nRows with multiple accession numbers:")
display(rows_with_commas)
print(f"\nNumber of rows with multiple accession numbers: {rows_with_commas.shape[0]}")
print('total number of rows in dataset before splitting:', carlino_supp.shape[0])

carlino_supp['Accession number'] = carlino_supp['Accession number'].str.split(',')
carlino_supp = carlino_supp.explode('Accession number')
# strip white spaces in the 'Accession number' column
carlino_supp['Accession number'] = carlino_supp['Accession number'].str.strip()

end_rows = carlino_supp.shape[0]
print('number of rows at the end:', end_rows)

```

```python
# Add Reference column from carlino_supp to carlino_metadata
# Create a dictionary from carlino_supp, keeping the first occurrence of each project_accession
reference_dict = dict(zip(carlino_supp['Accession number'], carlino_supp['Reference']))
carlino_metadata['Reference'] = carlino_metadata['project_accession'].map(reference_dict)

# Verify the merge by checking if References were added
print("Number of rows with References:", carlino_metadata['Reference'].notna().sum())
print("Total rows in carlino_metadata:", len(carlino_metadata))
```

```python
# checking the missing references
missing_refs = carlino_metadata[
    (carlino_metadata.Reference.isna()) & 
    (carlino_metadata['fermented/non-fermented'] == 'F')
][['dataset_name', 'project_accession','Reference']].drop_duplicates()
display(missing_refs)
```

```python
# saving the missing references table to a file
# will locate references manually
missing_refs.to_csv('data/missing_references.tsv', sep='\t', index=False)
```

```python
# Split source column into dataset_name and sample_id
dataset[['dataset_name', 'sample_id']] = dataset['mag_id'].str.rsplit('__', n=2, expand=True).iloc[:, :2]
dataset.head()
```

```python
# Check for duplicates in carlino_metadata using dataset_name and sample_id as keys
duplicates = carlino_metadata[carlino_metadata.duplicated(['dataset_name', 'sample_id'], keep=False)]

# Display counts of duplicates if any exist
if len(duplicates) > 0:
    print(f"Found {len(duplicates)} duplicate entries")
    # Group duplicates and show their counts
    duplicate_counts = duplicates.groupby(['dataset_name', 'sample_id']).size().reset_index(name='count')
    display(duplicate_counts)
    
    # Display the actual duplicate rows
    print("\nDuplicate rows:")
    display(duplicates)
else:
    print("No duplicates found in carlino_metadata")
```

```python
# Drop duplicates keeping the first occurrence
carlino_metadata = carlino_metadata.drop_duplicates(['sample_id'], keep='first')

# Verify the duplicate is removed
print("Number of duplicates after removal:",
    carlino_metadata.duplicated(['dataset_name', 'sample_id']).sum())
```

```python
# check the number of unique values in each column of carlino_metadata
display(carlino_metadata.shape, carlino_metadata.nunique())
```

```python
# Merge dataset with selected columns from carlino_metadata
merged_df = pd.merge(dataset, 
                    carlino_metadata.iloc[:, :17], 
                    how='left',
                    on=['dataset_name', 'sample_id'])

# Verify the merge
print("\nShape after merge:")
display(merged_df.shape, dataset.shape)
```

```python
#print the unique values of columns study_catalog for data that was not found in carlino
# Get value counts of study_catalog for entries with missing sequencing_platform
study_counts = merged_df[merged_df['sequencing_platform'].isna()]['study_catalog'].value_counts().to_frame()
display(study_counts)
print(f"\nTotal number of unique study_catalog values for samples without sequencing platform: {len(study_counts)}")

# Print total number of rows with missing sequencing_platform
print(f"\nTotal samples without sequencing platform: {merged_df['sequencing_platform'].isna().sum()}")
```
